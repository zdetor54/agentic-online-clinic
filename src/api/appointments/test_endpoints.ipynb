# Testing FastAPI Appointment Endpoints in a Notebook

This notebook demonstrates how to test your appointment API endpoints using the `requests` library. Make sure your FastAPI server is running (e.g., with `uvicorn`) before running these cells.

Endpoints covered:
- Create appointment (POST)
- Get appointment by ID (GET)
- List/search appointments (GET)
- Update appointment (PUT)
- Delete appointment (DELETE)

> Adjust the payloads and IDs as needed for your data model.

```python
# Import required libraries
import json

import requests

BASE_URL = "http://localhost:8000/appointments"  # Adjust if your server runs elsewhere
```

## 1. Create a new appointment (POST)
Replace the payload fields with valid values for your schema.

```python
# Example payload - update fields as needed
payload = {
    "patient_id": 8,  # Replace with a valid patient ID
    "appointment_date": "2026-02-15",
    "appointment_start_time": "10:00:00",
    "appointment_end_time": "10:30:00",
    "appointment_reason": "Routine checkup",
}

response = requests.post(BASE_URL + "/", json=payload, timeout=10)
print("Status code:", response.status_code)
try:
    print("Response:", response.json())
except Exception:
    print("No JSON response.")
```

## 2. Get appointment by ID (GET)
Replace `appointment_id` with a valid ID returned from the previous cell.

```python
# Test GET /{appointment_id} - Get appointment by ID
appointment_id = 7  # Change this to test different IDs

response = requests.get(f"{BASE_URL}/{appointment_id}", timeout=10)
print(f"GET /appointments/{appointment_id}")
print(f"Status code: {response.status_code}")
print("-" * 50)

if response.status_code == 200:
    appointment = response.json()
    print("✓ Appointment found:")
    print(json.dumps(appointment, indent=2))
elif response.status_code == 404:
    print("✗ Appointment not found")
    print(response.json())
else:
    print(f"✗ Unexpected status code: {response.status_code}")
    try:
        print(response.json())
    except Exception:
        print(response.text)
```

## 3. List/search appointments (GET)
You can filter by date or patient_id, or get all appointments.

```python
# Test GET / - List/search appointments

print("=" * 50)
print("Test 1: Get all appointments")
print("=" * 50)
response = requests.get(BASE_URL + "/", timeout=10)
print(f"Status code: {response.status_code}")
if response.status_code == 200:
    appointments = response.json()
    print(f"✓ Found {len(appointments)} appointment(s)")
    print(json.dumps(appointments, indent=2))
else:
    print(f"✗ Error: {response.status_code}")
    print(response.json())

print("\n" + "=" * 50)
print("Test 2: Filter by appointment_date")
print("=" * 50)
params = {"appointment_date": "2026-02-15"}
response = requests.get(BASE_URL + "/", params=params, timeout=10)
print(f"Status code: {response.status_code}")
if response.status_code == 200:
    appointments = response.json()
    print(f"✓ Found {len(appointments)} appointment(s) on 2026-02-15")
    print(json.dumps(appointments, indent=2))
else:
    print(f"✗ Error: {response.status_code}")
    print(response.json())

print("\n" + "=" * 50)
print("Test 3: Filter by patient_id")
print("=" * 50)
params = {"patient_id": 8}
response = requests.get(BASE_URL + "/", params=params, timeout=10)
print(f"Status code: {response.status_code}")
if response.status_code == 200:
    appointments = response.json()
    print(f"✓ Found {len(appointments)} appointment(s) for patient 8")
    print(json.dumps(appointments, indent=2))
else:
    print(f"✗ Error: {response.status_code}")
    print(response.json())

print("\n" + "=" * 50)
print("Test 4: Filter by both date and patient_id")
print("=" * 50)
params = {"appointment_date": "2026-02-15", "patient_id": 8}
response = requests.get(BASE_URL + "/", params=params, timeout=10)
print(f"Status code: {response.status_code}")
if response.status_code == 200:
    appointments = response.json()
    print(f"✓ Found {len(appointments)} appointment(s) for patient 8 on 2026-02-15")
    print(json.dumps(appointments, indent=2))
else:
    print(f"✗ Error: {response.status_code}")
    print(response.json())
```

## 4. Update an appointment (PUT)
Replace `appointment_id` and payload fields as needed.

```python
# Test PUT /{appointment_id} - Update appointment
appointment_id = 5  # Replace with a valid appointment ID

update_payload = {
    "patient_id": 8,
    "appointment_date": "2026-02-15",  # Date only, no time component
    # "appointment_start_time": "14:00:00",
    # "appointment_end_time": "14:30:00",
    "appointment_reason": "Follow-up checkup",
    "updated_by": "admin",
}

response = requests.put(f"{BASE_URL}/{appointment_id}", json=update_payload, timeout=10)
print(f"PUT /appointments/{appointment_id}")
print(f"Status code: {response.status_code}")
print("-" * 50)

if response.status_code == 200:
    appointment = response.json()
    print("✓ Appointment updated successfully:")
    print(json.dumps(appointment, indent=2))
elif response.status_code == 404:
    print("✗ Appointment not found")
    print(response.json())
else:
    print(f"✗ Error: {response.status_code}")
    try:
        print(response.json())
    except Exception:
        print(response.text)
```

## 5. Delete an appointment (DELETE)
Replace `appointment_id` with the ID to delete.

```python
appointment_id = 10  # Replace with a valid appointment ID
response = requests.delete(f"{BASE_URL}/{appointment_id}", timeout=10)
print("Status code:", response.status_code)
if response.status_code == 204:
    print("Appointment deleted successfully.")
else:
    try:
        print("Response:", response.json())
    except Exception:
        print("No JSON response.")
```
